```python
class Ders:
    def __init__(self, ders_numarasi):
        self.ders_numarasi = ders_numarasi
        self.ders_notu = self.notu_al()
        self.agirlik = self.agirlik_al()

    def notu_al(self):
        while True:
            try:
                ders_notu = int(input(f"{self.ders_numarasi} için ders notunu giriniz: "))
                if ders_notu < 0 or ders_notu > 100:
                    print("Ders notu 0-100 arasında olmalıdır.")
                    continue
                return ders_notu
            except ValueError:
                print("Lütfen bir tam sayı giriniz.")

    def agirlik_al(self):
        while True:
            try:
                agirlik = int(input(f"{self.ders_numarasi} için ağırlık notunu giriniz: "))
                if agirlik < 0 or agirlik > 10:
                    print("Ağırlık notu 0-10 arasında olmalıdır.")
                    continue
                return agirlik
            except ValueError:
                print("Lütfen bir tam sayı giriniz.")


class Öğrenci:
    def __init__(self):
        self.dersler = []
        self.ders_sayisi = self.ders_sayisini_al()
        self.dersleri_ekle()

    def ders_sayisini_al(self):
        while True:
            try:
                ders_sayisi = int(input("Ders sayısını giriniz: "))
                if ders_sayisi <= 0:
                    print("Ders sayısı pozitif bir tam sayı olmalıdır.")
                    continue
                return ders_sayisi
            except ValueError:
                print("Lütfen bir tam sayı giriniz.")

    def dersleri_ekle(self):
        for i in range(1, self.ders_sayisi + 1):
            self.dersler.append(Ders(i))

    def gano_hesapla(self):
        toplam = 0
        toplam_agirlik = 0
        for ders in self.dersler:
            toplam += ders.ders_notu * ders.agirlik
            toplam_agirlik += ders.agirlik
        return toplam / toplam_agirlik if toplam_agirlik != 0 else 0


class HarfNotu:
    @staticmethod
    def harf_notu_yazdir(gano):
        print("GANO: ", gano)
        print("HARF NOTU: ", end="")
        if gano >= 85:
            print("AA")
        elif gano >= 70:
            print("BA")
        elif gano >= 55:
            print("BB")
        elif gano >= 40:
            print("CB")
        else:
            print("FF")


# Kullanım
ogrenci = Öğrenci()
gano = ogrenci.gano_hesapla()
HarfNotu.harf_notu_yazdir(gano)
```
